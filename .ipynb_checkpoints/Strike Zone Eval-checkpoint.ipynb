```python
def postgame_report(position,
                    tagged = 'TaggedPitchType',
                    pitchcall = 'PitchCall',
                    pitcher = 'Pitcher',
                    batter = 'Batter',
                    PlateLocHeight = 'PlateLocHeight',
                    PlateLocSide = 'PlateLocSide'
                    ):
    
    # ---------- Select data ---------
    if position == 'pitching' : data = 'Spring 2021 - All Data Pitching.csv'
    elif position == 'hitting' : data = 'Spring 2021 - All Data Hitting.csv'
 
    # ---------- Import Libraries ----------
    import os
    import csv
    import pandas as pd
    import numpy as np
    import seaborn as sns
    import matplotlib.pyplot as plt
    from PyPDF2 import PdfFileMerger
    from scipy.interpolate import griddata
    sns.set_style('whitegrid')

    # ---------- Import CSV and Clean ----------
    df = pd.read_csv(data).replace(0, np.nan).round(2)
    df = df[df[tagged] != 'Undefined']
    
    # --------- Edit based on position --------
    if position == 'pitching' : pos = pitcher
    elif position == 'hitting' : pos = batter   
        
    if position == 'pitching' : index = df[pitcher].unique().tolist()
    elif position == 'hitting' : index = df[batter].unique().tolist()
    
    count = 0
    for name in index:
        # Set x and y coordinates
        try: 
            df_heatmap = df[df[pitchcall] == 'StrikeSwinging']
            df_heatmap = df_heatmap[df_heatmap[pos] == name]
            y = df_heatmap['PlateLocHeight'].dropna() 
            x = df_heatmap['PlateLocSide'].dropna()

            heatmap, xedges, yedges = np.histogram2d(x, y, bins=5)
            extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]

            plt.clf()
            plt.imshow(heatmap.T, extent=extent, origin='lower', cmap = 'coolwarm')
            plt.colorbar()

            #Lables
            plt.title('Strike Swinging Heat Map \n ' + name)
            plt.xlabel('<L>     Side in Feet     <R>')
            plt.ylabel('Height in Feet')

            # Chart dimensions
            plt.xlim([-1.5, 1.5])
            plt.ylim([0.5, 4.5]) 

            # Creates Strikezone lines
            plt.axhline(y=1.5, linewidth=2, color='k')
            plt.axhline(y=3.6, linewidth=2, color='k')
            plt.axvline(x=(2/3), linewidth=2, color='k')
            plt.axvline(x=(-2/3), linewidth=2, color='k')

            plt.savefig(name + '.pdf', bbox_inches='tight')
            plt.show()
            count += 1
        
        except:
            pass
        
    merger = PdfFileMerger()
    for name in index:
        try:merger.append(name + '.pdf')
        except: pass
    merger.write(position + " result.pdf")
    merger.close()

    for name in index:  
        os.remove(name + '.pdf')        
```

```python
postgame_report(position = 'pitching')
```
