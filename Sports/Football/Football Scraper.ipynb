```python
import requests
import bs4
from bs4 import BeautifulSoup as soup
import re
import pandas as pd
import numpy as np
```

```python
base_url = 'https://www.espn.com/nfl/playbyplay/_/gameId/'
game_code = 401220225
url = str(base_url) + str(game_code)
```

```python
html = requests.get(url)
soup = bs4.BeautifulSoup(html.content, 'html.parser')
new_soup = soup.find('div',attrs = {'id':'gamepackage-drives-wrap'})
```

```python
time = []
quarter = []
formation = []
play = []
```

```python
events = re.findall('\(\d{1,2}:\d{1,2}\W\-\W\d\D\D\).{1,300}', str(new_soup))
d_and_d = re.findall('\d\w{2}\W\&\w{3}\;\W\d{1,2}\W\w{2}\W\w{3}\W\d{1,2}', str(new_soup))

for event in events:
    # List of times
    times = str(re.findall('\d{1,2}:\d{1,2}', str(event)))
    time.append(times[2:-2])
    
    # Quarter
    quarters = str(re.findall('\-\W\d\D\D', str(event)))
    quarter.append(quarters[3:7])
    
    # list of formations
    formations = str(re.findall('\(\w{1,15}\)', str(event)))
    formation.append(formations[3:-3])
    
    # List of Plays
    plays = str(re.findall('\)\W.{1,300}', str(event)))
    print(plays)
```

```python
# Dataframe information
list_of_col = list(zip(quarter, time, formation))
name_of_col = ['Quarter', 'Time', 'Formation']

# Create Dataframe and check it
play_by_play = pd.DataFrame(list_of_col, columns = name_of_col)
play_by_play
```
